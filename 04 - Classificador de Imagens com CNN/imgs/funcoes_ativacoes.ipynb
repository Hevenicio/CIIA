```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = np.linspace(-5, 5, 100)
```

```python
def Sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def Tanh(x):
    return np.tanh(x)
```

```python
def ReLU(x):
    return np.maximum(0, x)
```

```python
# ... plota os valores da função ReLU
fig, ax = plt.subplots(figsize=(8, 5), dpi=100, facecolor='#00FA9A')
ax.plot(x, ReLU(x), color='k', lw=1)

# Configurações do gráfico
ax.set_title('ReLU', fontsize = 15, color = 'k')
ax.set_facecolor('#00FA9A')
ax.grid(True, color = 'white', lw = 0.8, linestyle = '--')
ax.set_xlabel('x', fontsize = 12)
ax.set_ylabel('ReLU(x)', fontsize = 12)
ax.spines[['top', 'right']].set_visible(False)
```

```python
fig, ax = plt.subplots(1, 3, figsize = (12, 4), dpi = 100, facecolor = '#00FA9A')

# ----- Lista de funções -----
ativacoes = [
    (Sigmoid, 'Sigmoid', 'Sigmoid(x)'),
    (Tanh, 'Tanh', 'Tanh(x)'),
    (ReLU, 'ReLU', 'ReLU(x)')
]

# ----- Plotando -----
for i, (func, nome, ylabel) in enumerate(ativacoes):
    ax[i].plot(x, func(x), color = 'k', lw = 1)
    ax[i].set_title(nome, fontsize = 15, color = 'k')
    ax[i].set_xlabel('x', fontsize = 12)
    ax[i].set_ylabel(ylabel, fontsize = 12)
    ax[i].set_facecolor('#00FA9A')
    ax[i].grid(True, color = 'white', lw = 0.8, linestyle = '--')
    ax[i].spines[['top', 'right']].set_visible(False)

# ----- Ajustes finais -----
plt.tight_layout()
```
